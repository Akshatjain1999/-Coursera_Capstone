# Scrapping the data

```python
from pandas.io.html import read_html
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wiki = read_html(link,attrs={'class':'wikitable'})
df=wiki[0]
```

```python
df['Borough'].replace('Not assigned',np.nan,inplace=True)
```

```python
df.dropna(subset=['Borough'],inplace=True)
```

```python
df['Neighborhood']=df['Neighborhood'].str.replace('/',',')
```

# Adding latitude and longitude

```python
lat_lng = pd.read_csv('Geospatial_Coordinates.csv')
lat_lng.head()
```

```python
california_df=pd.merge(df,lat_lng,left_on='Postal code',right_on='Postal Code')
```

```python
california_df.drop('Postal Code',axis=1,inplace=True)
```

```python
california_df.head()
```

# generating map to visualize neighborhood

```python
toronto_df=california_df[california_df['Borough'].str.contains("Toronto")]
```

```python
toronto_df.shape
```

### Getting lat lng of toronto

```python
import geopy
import folium
```

```python
address ='Toronto'
geolocator = geopy.Nominatim()
location = geolocator.geocode(address)
```

```python
latitude=location.latitude
longitude = location.longitude
```

## Creating map of toronto with neighborhoods

```python
map_toronto = folium.Map(location=[latitude,longitude],zoom_start=10)
for lat , lng , borough , neighborhood in zip(california_df['Latitude'],california_df['Longitude'],california_df['Borough'],california_df['Neighborhood']):
    label = '{}, {}'.format(neighborhood,borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius = 5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.5,).add_to(map_toronto)
```
